```python
df2 = df.drop('customerID', axis=1)

df2['TotalCharges'] = pd.to_numeric(
    df2['TotalCharges'],
    errors='coerce'
)

df_encoded = pd.get_dummies(df2, drop_first=True)
```

```python
df_encoded.shape
```

```python
corr = df_encoded.corr()
```

```python
plt.figure(figsize=(8,10))

sns.heatmap(
    corr[['Churn_Yes']].sort_values(
        by='Churn_Yes',
        ascending=False
    ),
    annot=True,
    cmap='coolwarm'
)

plt.title('Features Correlated with Churn')
plt.show()
```

```python
df_clean = df.copy()
```

```python
df_clean.isnull().sum()
```

```python
df_clean['TotalCharges'] = pd.to_numeric(
    df_clean['TotalCharges'],
    errors='coerce'
)
```

```python
df_clean['TotalCharges'].isnull().sum()
```

```python
df_clean['TotalCharges'].fillna(
    df_clean['TotalCharges'].median(),
    inplace=True
)
```

```python
df_clean.isnull().sum()
```

```python
df_clean.drop(
    'customerID',
    axis=1,
    inplace=True
)
```

```python
df_clean.columns
```

```python
df_clean['Churn'] = df_clean['Churn'].map({
    'No':0,
    'Yes':1
})
```

```python
df_clean['Churn'].value_counts()
```

```python
df_encoded = pd.get_dummies(
    df_clean,
    drop_first=True
)
```

```python
df_encoded.shape
```

```python
df_encoded.info()
```

```python
df_encoded.to_csv(
    'Telco_Cleaned.csv',
    index=False
)
```

```python
total_customers = len(df)

churn_rate = (
    df_clean['Churn'].mean()*100
)

avg_tenure = (
    df_clean['tenure'].mean()
)

avg_monthly = (
    df_clean['MonthlyCharges'].mean()
)

print("Total Customers:", total_customers)
print("Churn Rate:", round(churn_rate,2))
print("Average Tenure:", round(avg_tenure,2))
print("Average Monthly Charges:", round(avg_monthly,2))
```
