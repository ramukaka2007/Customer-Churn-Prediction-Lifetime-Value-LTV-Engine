```python
df["EstimatedLTV"] = df["MonthlyCharges"] * df["tenure"]
```

```python
df[["MonthlyCharges","tenure","EstimatedLTV"]].head()
```

```python
df["RevenueSegment"] = pd.qcut(
    df["EstimatedLTV"],
    q=4,
    labels=["Bronze","Silver","Gold","Platinum"]
)
```

```python
df["RevenueSegment"].value_counts()
```

```python
df["CustomerValueScore"] = (
    df["TotalServices"] * 2
    + df["tenure"] / 12
    + df["MonthlyCharges"] / 10
)
```

```python
df[["CustomerValueScore"]].head()
```

```python
df["HighValueCustomer"] = np.where(
    df["CustomerValueScore"] >= df["CustomerValueScore"].median(),
    1,
    0
)
```

```python
df["HighValueCustomer"].value_counts()
```

```python
df.to_csv("Telco_LTV_FeatureEngineered.csv", index=False)
```
