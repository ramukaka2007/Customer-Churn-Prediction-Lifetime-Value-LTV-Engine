```python
df['AvgChargePerMonth'] = df['TotalCharges'] / (df['tenure'] + 1)
```

```python
df[['MonthlyCharges','TotalCharges','tenure','AvgChargePerMonth']].head()
```

```python
def tenure_group(tenure):
    if tenure <= 12:
        return "New"
    elif tenure <= 36:
        return "Regular"
    elif tenure <= 60:
        return "Loyal"
    else:
        return "Very Loyal"

df['TenureGroup'] = df['tenure'].apply(tenure_group)
```

```python
df['TenureGroup'].value_counts()
```

```python
def charge_category(charge):
    if charge < 35:
        return "Low"
    elif charge < 70:
        return "Medium"
    else:
        return "High"

df['ChargeCategory'] = df['MonthlyCharges'].apply(charge_category)
```

```python
df['ChargeCategory'].value_counts()
```

```python
service_columns = [
    'PhoneService_Yes',
    'MultipleLines_Yes',
    'OnlineSecurity_Yes',
    'OnlineBackup_Yes',
    'DeviceProtection_Yes',
    'TechSupport_Yes',
    'StreamingTV_Yes',
    'StreamingMovies_Yes'
]

df['TotalServices'] = df[service_columns].sum(axis=1)

df[['TotalServices']].head()
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
df['HighValueCustomer'] = (
    (df['MonthlyCharges'] > 70) &
    (df['tenure'] > 24)
).astype(int)
```

```python
df['HighValueCustomer'].value_counts()
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
from xgboost import XGBClassifier
```

```python
df = pd.read_csv("Telco_FeatureEngineered.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr = LogisticRegression(
    random_state=42,
    max_iter=1000
)

lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
print("Accuracy :", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr))
print("Recall   :", recall_score(y_test, y_pred_lr))
print("F1 Score :", f1_score(y_test, y_pred_lr))
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
cm = confusion_matrix(y_test, y_pred_lr)

print(cm)
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
print("Accuracy :", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall   :", recall_score(y_test, y_pred_rf))
print("F1 Score :", f1_score(y_test, y_pred_rf))
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
print(confusion_matrix(y_test, y_pred_rf))
```

```python
xgb = XGBClassifier(
    random_state=42,
    eval_metric="logloss"
)

xgb.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb.predict(X_test)
```

```python
print("Accuracy :", accuracy_score(y_test, y_pred_xgb))
print("Precision:", precision_score(y_test, y_pred_xgb))
print("Recall   :", recall_score(y_test, y_pred_xgb))
print("F1 Score :", f1_score(y_test, y_pred_xgb))
```

```python
print(classification_report(y_test, y_pred_xgb))
```

```python
print(confusion_matrix(y_test, y_pred_xgb))
```

```python
comparison = pd.DataFrame({

    "Model":[
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],

    "Accuracy":[
        accuracy_score(y_test,y_pred_lr),
        accuracy_score(y_test,y_pred_rf),
        accuracy_score(y_test,y_pred_xgb)
    ],

    "Precision":[
        precision_score(y_test,y_pred_lr),
        precision_score(y_test,y_pred_rf),
        precision_score(y_test,y_pred_xgb)
    ],

    "Recall":[
        recall_score(y_test,y_pred_lr),
        recall_score(y_test,y_pred_rf),
        recall_score(y_test,y_pred_xgb)
    ],

    "F1 Score":[
        f1_score(y_test,y_pred_lr),
        f1_score(y_test,y_pred_rf),
        f1_score(y_test,y_pred_xgb)
    ]

})

comparison
```
